# Logik

```python
functions = {lambda a,b : a & b, lambda a,b : a | b}
names = {"UND", "ODER"}
named_functions = dict(zip(names, functions))

def wahrheitstafel(logik):
    name, lf = logik

    boolean = (True, False)
    print(f'  {name:5} |  True   False')
    print(f'-----------------------------')
    for a in boolean:
        print(f' {a!s} \t|', end = "")
        for b in boolean:
             print(f'  {lf(a, b)!s:5}', end = "")
        print()

        

for lf in named_functions.items():
    wahrheitstafel(lf)
    print()
```
