```python
def gallery(images, rows, cols, cmap=None):
    for n in range(rows * cols):
        plt.subplot(rows, cols,n+1)
        plt.imshow(images[n], cmap) 
        ax = plt.gca()
        ax.axes.xaxis.set_visible(False)
        ax.axes.yaxis.set_visible(False)        
```

## Bunches

Die `load`- und `fetch`-Methoden liefern keine `numpy`-Arrays zurück wie die `make`-Methoden, sondern `Bunch`-Objekte. Diese enthalten neben den Daten noch zusätzliche Informationen, die wie bei einem Dictionary über Keys abgerufen werden können.

```python
iris = ds.load_iris()
print(f'Typ \t= {type(iris)}\nKeys \t= {iris.keys()}')
```

`Bunch`ist eine Unterklasse von `Dictionary`mit dem Vorteil, dass die Werte direkt als Attribute zur Verfügung stehen. So lassen sich auch die benötigten Daten und Targets erhalten:

```python
X, y = iris.data, iris.target
print(X[:5],'\n',y[:5])
```

Alternativ kann man auch direkt nur die X und y laden, zusätzlich die gesamte Struktur als `Pandas` `Dataframe`

```python
X, y = ds.load_iris(as_frame=True, return_X_y=True)
print(X[:5])
```

Die Bunches enthalten meist noch eine interessante Beschreibung:

```python
print(iris.DESCR[:1000])
```

## Digits
