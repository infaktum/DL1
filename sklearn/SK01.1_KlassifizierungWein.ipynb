```python
import numpy as np
import matplotlib.pyplot as plt

plt.set_cmap('prism');
```

```python
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler, KBinsDiscretizer
from sklearn.compose import ColumnTransformer
```

```python
wine = load_wine()
X, y = load_wine(as_frame=True, return_X_y=True)

columns = ["alcohol", "malic_acid", "ash", "alcalinity_of_ash", "magnesium", "total_phenols", "flavanoids",  \
"nonflavanoid_phenols", "proanthocyanins", "color_intensity", "hue", "od280/od315_of_diluted_wines", "proline"  ]
```

```python
preprocessor = ColumnTransformer(
    [
        ("scaler", StandardScaler(), columns)
    ],
    verbose_feature_names_out=False,
).set_output(transform="pandas")

X_out = preprocessor.fit_transform(X)
X_out.head(10)
```

```python
a1,a2 = 0,1

_, ax = plt.subplots()
scatter = ax.scatter(wine.data[:, a1], wine.data[:, a2], c=wine.target)
ax.set(xlabel=wine.feature_names[a1], ylabel=wine.feature_names[a2])
_ = ax.legend(scatter.legend_elements()[0], wine.target_names, loc="lower right", title="Classes" )
```
