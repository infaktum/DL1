# MNIST Klassifizierung mit Tensorflow

```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt
from ki_kurs.mnist import MnistDataset 
```

```python
# Laden des Datensatzes


import ssl
ssl._create_default_https_context = ssl._create_unverified_context

train_data = MnistDataset('daten/mnist_train.csv')
test_data  = MnistDataset('daten/mnist_test.csv')

train_images, train_labels = train_data[1:], train_data[0]
test_images, test_labels = test_data[1:], test_data[0]
```

```python
# Vorverarbeitung der Trainingsdaten
train_images = train_images.reshape((-1, 28 * 28))
train_images = train_images.astype('float32') / 255

# Vorverarbeitung der Testdaten
test_images = test_images.reshape((-1, 28 * 28))
test_images = test_images.astype('float32') / 255
```

```python
# Konvertierung der Labels in One-Hot-Kodierung
train_labels = tf.keras.utils.to_categorical(train_labels)
test_labels = tf.keras.utils.to_categorical(test_labels)
```

```python
# Aufbau des neuronalen Netzes
model = tf.keras.Sequential([
    tf.keras.layers.Dense(512, activation='relu', input_shape=(28 * 28,)),
    tf.keras.layers.Dense(10, activation='softmax')
])
```

```python
# Kompilieren des Modells
model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Training des Modells
model.fit(train_images, train_labels, epochs=5, batch_size=128)
```

```python
# Bewertung des Modells
test_loss, test_acc = model.evaluate(test_images, test_labels)
print('Test accuracy:', test_acc)
```
