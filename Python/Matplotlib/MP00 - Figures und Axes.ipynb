```python
import matplotlib.pyplot as plt

%matplotlib inline
```

### Einige Testdaten

```python
x,y = [1,2,3,4],[0,5,3,8]
```

[matplotlib Starter](https://matplotlib.org/stable/users/explain/quick_start.html)

Grafiken werden in `matplotlib` durch Objekte vom Type `Figure` und `Axes`erzeugt und manipuliert.

## Figure

Ein Figure-Objekt enthält eine vollständige Grafik mit einer oder mehreren Axes-Objekten. Es ist im wesentlichen ein _Container_-Objekt.

## Axes

Die eigentlichen Grafik-Operationen werden durch die Axes-Objekte durchgeführt.

Ein Axes-Objekt kann dabei sogar ein Container für weiter Figure-Objekte sein.

## Beispiele

### Figure mit einem einzelnen Axes-Objekt

```python
fig, ax = plt.subplots(figsize=(6,4), layout='constrained')
fig.suptitle("Figure.suptitle()")

ax.set_title("Axes.set_title()")
ax.plot(x,y)
pass
```

### Figure mit mehreren Axes-Objekten

```python
fig2, _ = plt.subplots(2,3,figsize=(20, 8))
fig.suptitle("Figure.suptitle()")

for a in fig2.get_axes():
    ax.plot(x,y)
    ax.set_title("Axes.set_title()")
```

```python
def my_plotter(ax, data1, data2, param_dict):
    """
    A helper function to make a graph.
    """
    out = ax.plot(data1, data2, **param_dict)
    return out
```
