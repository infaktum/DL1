# EX-03 Multi-Layer Perzeptron

* Ein Multi-Layer Perzeptron ist eine spezielle Art von Feedforward-neuronalen Netzwerk.
* Es besteht aus mindestens drei Schichten: einer Eingabeschicht, einer oder mehreren verborgenen Schichten und einer Ausgabeschicht.
* MLPs verwenden vollständig verbundene Schichten (d. h. jedes Neuron einer Schicht ist mit jedem Neuron der nächsten Schicht verbunden).
* Sie sind ein einfaches, aber leistungsfähiges Modell für Probleme wie die Klassifikation und Regression.
* MLPs verwenden eine nichtlineare Aktivierungsfunktion (z. B. ReLU, Sigmoid oder Tanh) in den verborgenen Schichten, um komplexe Muster zu lernen.
