```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return x * (1 - x)

class NeuralNetwork:
    def __init__(self, layers):
        self.layers = layers
        self.weights = []
        for i in range(1, len(layers)):
            self.weights.append(np.random.randn(layers[i], layers[i-1]))
        
    def feedforward(self, inputs):
        activations = inputs
        for w in self.weights:
            activations = sigmoid(np.dot(w, activations))
        return activations
    
    def train(self, inputs, targets, epochs, learning_rate):
        for epoch in range(epochs):
            for i in range(len(inputs)):
                input_data = inputs[i]
                target_data = targets[i]
                
                # Forward pass
                activations = [input_data]
                for w in self.weights:
                    activations.append(sigmoid(np.dot(w, activations[-1])))
                
                # Backpropagation
                error = target_data - activations[-1]
                deltas = [error * sigmoid_derivative(activations[-1])]
                for l in range(len(self.layers) - 2, 0, -1):
                    deltas.append(np.dot(self.weights[l].T, deltas[-1]) * sigmoid_derivative(activations[l]))
                deltas.reverse()
                
                # Update weights
                for l in range(len(self.weights)):
                    self.weights[l] += learning_rate * np.outer(deltas[l], activations[l]).T
    
# Beispiel-Daten
inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
targets = np.array([[0], [1], [1], [0]])

# Neuronales Netzwerk erstellen
network = NeuralNetwork([2, 3, 1])

# Trainieren des Netzwerks
network.train(inputs, targets, epochs=10000, learning_rate=0.1)

# Testen des trainierten Netzwerks
for i in range(len(inputs)):
    print("Input:", inputs[i], "Predicted Output:", network.feedforward(inputs[i]))
```
